```python
# The Aim: To automate the creative brainstorming process by turning a single raw sentence into a structured,
# ready-to-draw digital content script.
```

```python
user_input = " Adulthood is just being tired constantly, buying expensive coffee to stay awake, and then complaining you have no money."

comic_beats = user_input.split(',')

print("ðŸŽ¬ --- 3-PANEL STICK-FIGURE COMIC SCRIPT --- ðŸŽ¬\n")

print(f"PANEL 1:")
print(f"Visual: Stick figure looks exhausted at a desk.")
print(f"Caption: '{comic_beats[0].strip().capitalize()}...'")
print("-" * 40)

print(f"PANEL 2:")
print(f"Visual: Stick figure holding an oversized coffee cup.")
print(f"Caption: '{comic_beats[1].strip().capitalize()}...'")
print("-" * 40)

print(f"PANEL 3:")
print(f"Visual: Stick figure looking directly at the reader, checking an empty wallet.")
print(f"Caption: '{comic_beats[2].strip().capitalize()}...'")
print("-" * 40)
```

```python
from openai import OpenAI
from google.colab import userdata

secure_key = userdata.get('OPENAI_API_KEY')

client = OpenAI(api_key=secure_key)

print("ðŸ§  Asking the AI for a bitter-truth observation...\n")

response = client.chat.completions.create(
    model ="gpt-3.5-turbo",
    messages=[
        {"role":"system","content":"You are a cynical comedy writer who creates stick-figure comics."},
        {"role":"user","content": "Generate a single, one-sentence sarcastic, bitter-truth observation about daily life. It must have exactly three distinct clauses separated by commas."}
    ],

    temperature=0.8
)

ai_generated_sentence = response.choices[0].message.content
print(f"RAW AI OUTPUT: {ai_generated_sentence}\n")
print("ðŸŽ¬ --- AUTOMATED 3-PANEL SCRIPT --- ðŸŽ¬\n")

comic_beats = ai_generated_sentence.split(',')

if len(comic_beats) >= 3:
    print(f"PANEL 1:")
    print(f"Visual: Stick figure looks exhausted.")
    print(f"Caption: '{comic_beats[0].strip().capitalize()}...'")
    print("-" * 40)

    print(f"PANEL 2:")
    print(f"Visual: Stick figure attempts to fix the problem.")
    print(f"Caption: '{comic_beats[1].strip().capitalize()}...'")
    print("-" * 40)

    print(f"PANEL 3:")
    print(f"Visual: Stick figure staring blankly at the reader.")
    print(f"Caption: '{comic_beats[2].strip().capitalize()}.'")
else:
    print("The AI didn't use enough commas! Try running the script again.")

```
