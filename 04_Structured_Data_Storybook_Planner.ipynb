```python
# 1. Our "Dummy" AI Output (A perfectly structured Python Dictionary)
# In a real app, the OpenAI API would generate this exact structure for us!

storybook_data = {
    "title": "The Little Robot Who Forgot How to Beep",
    "target_age": "3-5 years",
    "characters": ["Pip the Robot", "Gears the Owl", "Wrench the Dog"],
    "pages": [
        {"page_number": 1, "text": "Pip was a shiny blue robot, but his beep was broken.", "illustration_idea": "Pip looking sad next to a broken speaker."},
        {"page_number": 2, "text": "He asked Gears the Owl for help.", "illustration_idea": "Pip standing under a tree talking to a mechanical owl."},
        {"page_number": 3, "text": "Gears said, 'A beep comes from the heart, not the wires!'", "illustration_idea": "Gears pointing at Pip's glowing chest plate."}
    ]
}

print("âœ… Structured data loaded successfully!")
```

```python
book_title = storybook_data['title']
age_group = storybook_data['target_age']

print(f"ğŸ“š Book Title: {book_title}")
print(f"ğŸ‘¶ Target Audience: {age_group}")

character_count = len(storybook_data['characters'])
print(f"ğŸ¤– Number of Characters : {character_count}")
```

```python
print("\nğŸ“ --- STORYBOAD OUTLINE --- ğŸ“\n")

for page in storybook_data["pages"]:

  page_num = page['page_number']
  story_text = page['text']
  art_prompt = page['illustration_idea']

  print(f"PAGE {page_num}: ")
  print(f"ğŸ“– Text: {story_text}")
  print(f"ğŸ¨Art Idea: {art_prompt}")
  print("-"*40)
```
