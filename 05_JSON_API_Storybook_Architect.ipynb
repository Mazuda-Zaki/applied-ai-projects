```python
import json
from openai import OpenAI
from google.colab import userdata

# Secure API setup
secure_key = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=secure_key)

print("ğŸ§  Forcing the AI to write a structured storybook...\n")

# 2. THE API CALL
response =  client.chat.completions.create(
    model ="gpt-3.5-turbo",
    # It strictly forbids the AI from using conversational filler
    response_format = {"type": "json_object"},
    messages =[
        {"role": "system", "content": "You are a professional children's book author. You must strictly output your response in JSON format."},
        {"role": "user" , "content": """
        Write an outline for a children's book about a brave little toaster who gets lost in a big kitchen.
        You must use this exact JSON structure:
        {
          "title": "Book Title",
          "target_age": "Age Range",
          "characters": ["Character 1", "Character 2"],
          "pages": [
            {
              "page_number": 1,
              "text": "Story text for page 1",
              "illustration_idea": "Art prompt for page 1"
            }
          ]
        }
        Make the story exactly 3 pages long, with each page object having a 'page_number', 'text', and 'illustration_idea'.
        """}
    ],
    temperature= 0.8
)

# 3.Catch the raw text from the AI
raw_ai_text = response.choices[0].message.content

# 4. The Magic conversion (string -> python dictionary)
storybook_data = json.loads(raw_ai_text)

# 5. The output loop
print(f"ğŸ“š TITLE: {storybook_data['title']}")
print(f"ğŸ‘¶ TARGET AGE: {storybook_data['target_age']}")
print(f"ğŸ¤– CHARACTERS: {', '.join(storybook_data['characters'])}\n")

print("ğŸ“ --- STORYBOARD OUTLINE --- ğŸ“\n")

for page in storybook_data["pages"]:
    print(f"PAGE {page['page_number']}:")
    print(f"ğŸ“– Text: {page['text']}")
    print(f"ğŸ¨ Art Idea: {page['illustration_idea']}")
    print("-" * 40)
```
